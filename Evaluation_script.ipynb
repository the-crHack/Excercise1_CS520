```python
def evaluate_single_code(code_string, function_name, test_cases):
    """
    Evaluate one code sample against test cases
    """
    results = {'passed': 0, 'failed': 0, 'total': len(test_cases)}

    try:
        exec_globals = {}
        exec(code_string, exec_globals)
        user_function = exec_globals.get(function_name)

        if not user_function:
            return results

        for test_case in test_cases:
            try:
                namespace = {function_name: user_function, '__builtins__': {}}

                def custom_assert(condition, message=None):
                    if not condition:
                        raise AssertionError(message or "Assertion failed")
                namespace['assert'] = custom_assert

                exec(test_case, namespace)
                results['passed'] += 1
            except AssertionError:
                results['failed'] += 1
            except Exception:
                results['failed'] += 1

    except Exception:
        results['failed'] = len(test_cases)

    return results

def calculate_pass_at_k(code_results_list, k=1):
    """
    Calculate pass@k metric
    """
    if len(code_results_list) < k:
        return 0.0

    n = len(code_results_list)
    c = sum(1 for result in code_results_list if result['passed'] == result['total'])

    if n - c < k:
        return 1.0

    pass_at_k = 1.0 - comb(n - c, k) / comb(n, k)
    return pass_at_k


# Put all your solutions in a list (you'll need to add solutions for problems 3-10)
all_solutions = [
    problem1_solutions,
    problem2_solutions,
    problem3_solutions,
    problem4_solutions,
    problem5_solutions,
    problem6_solutions,
    problem7_solutions,
    problem8_solutions,
    problem9_solutions,
    problem10_solutions,

]

# 3. Evaluate each problem with its 5 solutions
final_results = []
all_test_results = []  # Store all individual test results for overall success rate

print("\nEVALUATING EACH PROBLEM WITH 5 SOLUTIONS:")
print("=" * 80)

for i, problem in enumerate(problems_data):
    print(f"\nProblem {i+1}: {problem['entry_point']}")
    print(f"Description: {problem['text']}")

    # Get the 5 solutions for this problem
    if i < len(all_solutions):
        solutions = all_solutions[i]
    else:
        print("No solutions provided for this problem, skipping...")
        continue

    # Evaluate all 5 solutions
    all_results = []
    for j, solution in enumerate(solutions):
        result = evaluate_single_code(solution, problem['entry_point'], problem['test_list'])
        all_results.append(result)
        all_test_results.append(result)  # Store for overall calculation
        print(f"  Solution {j+1}: {result['passed']}/{result['total']} tests passed")

    # Calculate pass@k metrics
    pass_at_1 = calculate_pass_at_k(all_results, k=1)
    pass_at_3 = calculate_pass_at_k(all_results, k=3)
    pass_at_5 = calculate_pass_at_k(all_results, k=5)

    problem_result = {
        'problem_id': i + 1,
        'task_id': problem['task_id'],
        'function_name': problem['entry_point'],
        'pass@1': pass_at_1,
        'pass@3': pass_at_3,
        'pass@5': pass_at_5,
        'solutions_tested': len(solutions)
    }

    final_results.append(problem_result)

    print(f"  Results: pass@1={pass_at_1:.2f}, pass@3={pass_at_3:.2f}, pass@5={pass_at_5:.2f}")

# Calculate overall success rate
def calculate_overall_success_rate(all_test_results):
    """
    Calculate overall success rate across all problems and solutions
    """
    total_tests = 0
    total_passed = 0

    for result in all_test_results:
        total_tests += result['total']
        total_passed += result['passed']

    if total_tests == 0:
        return 0.0

    return total_passed / total_tests

# Calculate overall metrics
overall_success_rate = calculate_overall_success_rate(all_test_results)

# Count how many solutions passed all tests
solutions_passed_all = sum(1 for result in all_test_results if result['passed'] == result['total'])
total_solutions = len(all_test_results)
solutions_success_rate = solutions_passed_all / total_solutions if total_solutions > 0 else 0

# Final summary
print("\n" + "=" * 80)
print("FINAL PASS@K RESULTS")
print("=" * 80)
print(f"{'Problem':<8} {'Function':<20} {'pass@1':<8} {'pass@3':<8} {'pass@5':<8}")
print("-" * 80)

for result in final_results:
    print(f"{result['problem_id']:<8} {result['function_name']:<20} "
          f"{result['pass@1']:<8.2f} {result['pass@3']:<8.2f} {result['pass@5']:<8.2f}")

# Calculate averages
if final_results:
    avg_pass_1 = np.mean([r['pass@1'] for r in final_results])
    avg_pass_3 = np.mean([r['pass@3'] for r in final_results])
    avg_pass_5 = np.mean([r['pass@5'] for r in final_results])

    print("-" * 80)
    print(f"{'AVERAGE':<8} {'':<20} {avg_pass_1:<8.2f} {avg_pass_3:<8.2f} {avg_pass_5:<8.2f}")

    # Overall success rates
    print("\n" + "=" * 80)
    print("OVERALL SUCCESS RATES")
    print("=" * 80)
    print(f"Overall Test Success Rate: {overall_success_rate:.2%}")
    print(f"  Total Tests: {sum(r['total'] for r in all_test_results)}")
    print(f"  Tests Passed: {sum(r['passed'] for r in all_test_results)}")
    print(f"  Tests Failed: {sum(r['failed'] for r in all_test_results)}")

    print(f"\nSolution Success Rate: {solutions_success_rate:.2%}")
    print(f"  Total Solutions: {total_solutions}")
    print(f"  Solutions Passed All Tests: {solutions_passed_all}")
    print(f"  Solutions Failed Some Tests: {total_solutions - solutions_passed_all}")

    print("=" * 80)
```
